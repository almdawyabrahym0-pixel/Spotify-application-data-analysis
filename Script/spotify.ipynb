```python
import pandas as pd
```

```python
all_time_df = pd.read_csv(r"C:\Users\K.Asus\Desktop\programing\Data_Analysis\TEST\داتاSpotify\archive\spotify_alltime_top100_songs.csv")
artists_df = pd.read_csv(r"C:\Users\K.Asus\Desktop\programing\Data_Analysis\TEST\داتاSpotify\archive\spotify_wrapped_2025_top50_artists.csv")
songs_2025_df = pd.read_csv(r"C:\Users\K.Asus\Desktop\programing\Data_Analysis\TEST\داتاSpotify\archive\spotify_wrapped_2025_top50_songs.csv")
```

```python
##تنظيف الو ملف اكثر  اغنيه استماعا 
all_time_df.head()
```

```python
all_time_df.info()
```

```python
# 1. إزالة أي صفوف مكررة بناءً على اسم الأغنية والفنان
all_time_df.drop_duplicates(subset=['song_title', 'artist'], inplace=True)

# 2. التعامل مع القيم المفقودة (Nulls)
# لو في قيم مفقودة في الأعمدة الرقمية ممكن نعوضها بـ 0 أو المتوسط
all_time_df['total_streams_billions'] = all_time_df['total_streams_billions'].fillna(0)

# 3. التأكد من نوع البيانات
all_time_df['release_year'] = all_time_df['release_year'].astype(int)

print("تم تنظيف ملف الأكثر استماعاً بنجاح.")
```

```python
##تتنظيف الملف الثاني افضل50 فنان لسنه2025 
artists_df.head()
```

```python
artists_df.info()
```

```python
##توحيد حالة الحروف وتنسيق النصوص (إزالة المسافات الزائدة)
artists_df['artist_name'] = artists_df['artist_name'].str.strip()
artists_df['primary_genre'] = artists_df['primary_genre'].str.title()

#  التأكد من أن قيم "جوائز غرامي" أرقام صحيحة
artists_df['grammy_wins'] = artists_df['grammy_wins'].fillna(0).astype(int)
print("تم تنظيف ملف اكثر 50 فنان لسنه 2025.")
```

```python
##تنظيف ملف اكثر 50 اغنيه استماع لسنه 2025
songs_2025_df.head()
```

```python
songs_2025_df.info()
```

```python
# 1. تحويل المدة من ثواني إلى دقائق (اختياري لتحسين القراءة)
songs_2025_df['duration_min'] = songs_2025_df['duration_seconds'] / 60

# 2. التعامل مع القيم المنطقية (Boolean)
# التأكد من أن عمود 'explicit' معالج بشكل صحيح
songs_2025_df['explicit'] = songs_2025_df['explicit'].astype(bool)

# 3. التحقق من القيم المتطرفة (Outliers) في الـ BPM
# مثلاً التأكد أن مفيش أغنية الـ BPM بتاعها صفر
songs_2025_df = songs_2025_df[songs_2025_df['bpm'] > 0]

print("تم تنظيف ملف أغاني 2025 بنجاح.")
```
